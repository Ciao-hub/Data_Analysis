```python
import matplotlib.pyplot as plt
import numpy as np
```

plot样式参数：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
y = [np.random.randint(0,10) for x in range(10)]
plt.plot(y, linewidth = 4, color = 'r', linestyle = '--')   # 设置线条的宽度和颜色
```

```python
# set_的方法设置线条样式，只能设置单条线的样式
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
line = lines[0]
line.set_color('g')
line.set_linewidth(3)
print(lines)
```

```python
# plt.setp的方式设置多条线条
```

```python
y1 = [np.random.randint(0,10) for x in range(10)]
y2 = [np.random.randint(0,10) for x in range(10)]
lines = plt.plot(range(10),y1,range(10),y2)
plt.setp(lines, linewidth = 3)
```

### 笔记：
1. 使用plot绘图的时候，就可以传递Line2D的属性值进去，来修改线条样式
2. 使用plot返回的线条，单独设置也可以，但是单独设置时，是通过'set_Line2D属性名()'的方式
3. 使用plot.setp来一次性设置多条线条的样式
