```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,20,50)  # 均匀间隔创建数值序列，linspace(start,stop,num)
y = np.sin(x)
plt.plot(x,y, marker = 'o', markerfacecolor = 'r', markersize = 8)
```

```python
# 文本注释
plt.plot(np.sin(np.arange(20)),marker = 'o')
plt.annotate("(0,0)",xy = (0,0),xytext = (-0.8,-0.5),
             arrowprops = {"width":2,'headwidth':5,'headlength':20,'shrink':0.1,'facecolor':'y'})
```

```python
a = [1,2,3,4]
for index,x in enumerate(a):
    print(index,x)
```

```python
y = np.sin(range(20))
plt.plot(y,marker = 'o')
for index,value in enumerate(y):
    plt.annotate("(%d,%.2f)"%(index,value),xy = (index,value),xytext = (index - 0.1, value - 0.1))
```

### 笔记：
1. plt.plot的时候，可以通过marker参数设置坐标点的样式。markerfacecolor代表点的颜色，markeredgecolor代表边界颜色，marker代表点的大小。
2. plt.annotate(text,xy,xytext,arrowprops)是用来做注释的，text代表需要注释的文本，xy代表需要注释的坐标点，xytext代表文本的坐标点，arrowprops代表箭头的一些属性
3. annotate官方文档：https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html?highlight=annotate#matplotlib.pyplot.annotate
