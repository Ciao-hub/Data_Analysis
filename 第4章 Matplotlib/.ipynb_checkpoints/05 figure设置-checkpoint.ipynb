```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize = (8,5),facecolor = 'g',edgecolor = 'k',linewidth = 2,dpi = 88)
plt.plot(np.arange(5,20))
plt.savefig('D:\\Workspace\\Data_Analysis\\第4章 Matplotlib\\a.png')
# ax = plt.gca()
# ax.set_facecolor('g')
```

### 笔记：
1. 画板的样式：可以调用plt.figure来实现，这个参数调用必须要在所有绘图之前完成。参数解释：
    * figsize：画板尺寸，是一个元组，第一个参数是宽度，第二个参数是高度，单位都是英寸
    * facecolor：画板颜色，注意区分Axes的背景颜色
    * edgecolor：边框颜色，边框默认宽度为0，如果需要看到边框，需要设置linewidth参数
    * dpi：像素，也就是一个英寸中的像素点个数。如果想要增加分辨率，那么dpi设置高点
2. 保存图片：
    * plt.savefig
    * 鼠标右键另存为
