```python
import matplotlib.pyplot as plt
import numpy as np
```

**绘制多根折线**

```python
x = np.linspace(0, 20)
plt.plot(x,np.sin(x),x,np.cos(x))
```

**绘制多个子图**

```python
values = np.arange(20)
plt.subplot(221)  # 221代表绘制2行2列中的第1个图
plt.plot(values)
plt.plot(values**2)
plt.subplot(222)
plt.plot(np.sin(values),'r')
plt.subplot(223)
plt.plot(np.cos(values),'b')
plt.subplot(224)
plt.plot(np.tan(values),'y')
```

```python
values = np.linspace(0,20)
plt.style.use('seaborn')  # 设置样式
fig,axes = plt.subplots(2,2,sharex = True, sharey = True)  # 2行2列
ax1 = axes[0,0]
ax1.plot(values)
ax2 = axes[0,1]
ax2.plot(np.sin(values),'r')
ax3 = axes[1,0]
ax3.plot(np.cos(values),'b')
ax4 = axes[1,1]
ax4.plot(np.tan(values),'y')
```

### 笔记：
1. 如果想要在一个图上绘制多根折线，那么直接在plot中传递多个x和y就可以了，或者调用多次plot方法就可以
2. 如果想要在一个画板上绘制多个子图（Axes对象），那么可以使用plt.subplot或plt.subplots
    * plt.subolot：plt.subolot(221) 代表2行2列第1个子图，使用subplot后绘制的所有东西都是在当前子图上，直到出现新的subplot
    * plt.subplots：fig,axes = plt.subplots(2,2,sharex = True)第一个参数代表行，第二个参数代表列，这个方法是一次性把所有子图都绘制了，但是子图上没有任何的东西，我们需要他的返回值来绘制，axes就是子图的对象。
