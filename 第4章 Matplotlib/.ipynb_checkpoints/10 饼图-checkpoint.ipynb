```python
oses = {
'windows7':60.86,
'windows10': 18.46,
'windows8': 3.61,
'windows xp': 10.3,
'mac os': 6.78,
'其他': 1.12
}
patches,texts,autotexts = plt.pie(oses.values(), autopct = "%.2f%%",labels = oses.keys(),
       textprops = {'fontproperties':font},shadow = True,
       explode = (0,0.1,0,0,0,0))  # 返回字典的所有值
for autotext in autotexts:
    autotext.set_color('w')
plt.show()
```

### 笔记：
1. 在`matplotlib`中，可以通过`plt.pie`来实现，其中的参数如下：
    * `x`：饼图的比例序列
    * `labels`：饼图上每个分块的名称文字。
    * `explode`：设置某几个分块是否要分离饼图。
    * `autopct`：设置比例文字的展示方式。比如保留几个小数等。
    * `shadow`：是否显示阴影。
    * `textprops`：文本的属性（颜色，大小等）。
2. 返回值:
    * `patches`：饼图上每个分块的对象。
    * `texts`：分块的名字文本对象。
    * `autotexts`：分块的比例文字对象。
