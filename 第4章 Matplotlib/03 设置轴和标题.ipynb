```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import font_manager   # 不支持中文，需要手动引入
```

```python
y = [np.random.randint(10) for x in range(10)]
font = font_manager.FontProperties(fname = r'C:\Windows\Fonts\msyh.ttc', size = 15)
plt.plot(y)
plt.title("折线图",fontproperties = font)
```

```python
avenger = [17974.4, 50918.4, 30033.0, 40329.1, 52330.2, 19833.3,11902.0]
plt.figure(figsize = (10, 5))   # 改变图表大小
plt.plot(avenger, marker = 'o')
font = font_manager.FontProperties(fname = r'C:\Windows\Fonts\msyh.ttc', size = 15)
plt.xticks(range(7),['第%d天'%x for x in range(1,8)],fontproperties = font)  # 设置横坐标
plt.xlabel("天数",fontproperties = font)  # x轴标签
 
plt.ylabel("票房（单位：万）",fontproperties = font)  # y轴标签
plt.title("复仇者联盟票房",fontproperties = font)   # 设置标题

plt.grid(True)  # 设置网格

plt.show()
```

### 笔记：
1. 标题通过plt.title来设置
2. 刻度通过xticks和yticks来设置，需要传递两个参数，第一个参数不重要，只要和数据量保持一致就可以，比如有20个数据，只要产生20个数据的数组或列表都可以，第二个参数label是用来设置每个刻度上的文本显示
3. 轴的名称是通过xlabel和ylabel来设置的
4. 默认plot不支持中文，如果想要支持中文，需要创建一个字体对象'font_manager.FontProperties'，然后指定fname参数，需要具体路径。拷贝出来的字体的路径，前面有个乱码，需要删掉。
