```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
y = [np.random.randint(0,10) for x in range(10)]
plt.plot(y)
```

```python
y = [np.random.randint(0,10) for x in range(20)]
x = range(5,25)
plt.plot(x,y)
```

```python
y1 = [np.random.randint(0,10) for x in range(20)]
x1 = range(5,25)
plt.plot(x1,y1)
```

```python
# 传入字典
mydata = {
    'a': range(5, 25),
    'b': [np.random.randint(0,10) for x in range(20)]
}
mydf = pd.DataFrame(data = mydata)
plt.plot('a','b',"",data = mydf)
```

```python
# 改变线条样式
mydata = {
    'a': range(5, 25),
    'b': [np.random.randint(0,10) for x in range(20)]
}
mydf = pd.DataFrame(data = mydata)
plt.plot('a','b',"^y",data = mydf)
```

### 笔记：
1. plt.plot可以只传输Y轴的值，如果只传Y轴的值，那么X轴就会默认使用range（0， Y的长度）
2. plt.plot的x和y参数不能够作为关键字参数来传输，只能作为位置参数来传
3. plt.plot中的data参数可以作为一个字典或者DataFrame对象，然后在x和y指定这个列的名字，那么plot会自动读取。一个细节，因为plt.plot后面对应三个参数，'x' 'y' 'fmt'， 如果只传入x和y，那么会产生歧异，可以多传一个参数，表示颜色
4. plt.plot的fmt参数可以设置线条的样式以及颜色
5. plt.plot的color参数可以使用字母、十六进制，或者RGBA的方式设置颜色
