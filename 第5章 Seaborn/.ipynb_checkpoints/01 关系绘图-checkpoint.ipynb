```python
import seaborn as sns
import pandas as pd
```

**基本使用**

```python
tips = sns.load_dataset('tips')  # 调用Github中seaborn-data项目，https://github.com/mwaskom/seaborn-data
# tips = pd.read_csv('tips.csv') # 也可下载到本地使用
tips
```

```python
sns.relplot(x = 'total_bill',y = 'tip', data = tips, size = 'size')
```

**hue参数**

```python
sns.relplot(x = 'total_bill',y = 'tip', hue = 'day', data = tips)  # 根据星期绘制颜色
```

**添加col和row参数**

```python
# 把'time'中的两项'Lunch' 'Dinner'分割成两个图显示
sns.relplot(x = 'total_bill',y = 'tip', col = 'time', data = tips)
```

```python
sns.relplot(x = 'total_bill',y = 'tip', row = 'time', data = tips)
```

```python
sns.relplot(x = 'total_bill',y = 'tip', col = 'time', row = 'sex',data = tips)
```

**指定具体多少列**

```python
# col_wrap参数指定列数
sns.relplot(x = 'total_bill', y = 'tip', col = 'day',col_wrap = 4,data = tips)
```

**绘制折线图**

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line',data = fmri)  # 蓝色透明部分表示置信区间
```

```python
sns.relplot(x = 'timepoint',y = 'signal',data = fmri, 
            kind = 'line', hue = 'region',style = 'region',col = 'event')
```

```python
sns.lineplot(x = 'timepoint', y = 'signal', data = fmri, hue = 'region', style = 'region')
```

```python
# 绘制简单折线图
import numpy as np
sns.relplot(x = 'x',y = 'y',
            data = pd.DataFrame({'x':np.arange(10),'y':np.sin(np.arange(10))}),kind = 'line')
```

### 笔记：
1. relplot是用来绘制关系数据图的，scatterplot和lineplot是其中的一个封装
2. replot绘制的时候，不能直接给x和y指定具体的值，而应该使用DataFrame中列的名字
3. hue表示的是颜色，应该是指定某个列的名字，会自动将指定的列的值的个数，取不同颜色
4. col表示分成几个图像，应该指定某个列的名字
5. style在绘制折线图的时候，可以用来指定线条的样式
6. 官方文档：https://seaborn.pydata.org/generated/seaborn.relplot.html?highlight=relplot#seaborn.relplot
