```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
axes = sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
axes.set_xticks(range(0,60,5))
plt.show()
```

```python
fig,[ax1,ax2] = plt.subplots(1,2,figsize = (10,5))
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips,ax = ax1)
sns.barplot(x = 'day', y = 'total_bill',data = tips,ax = ax2)
```

```python
sns.relplot(x = tips['total_bill'], y = tips['tip'])
```

### 笔记:
 1. FacetGrid，可以通过col和row等参数来一次性构建多个图形
 2. replot、catplot、lmplot等函数底层都是使用FacetGrid技术
 3. 凡是函数名中已经明确这个图的类型，都是直接使用Axes进行绘图
 4. FacetGrid绘图的x和y参数，必须为DataFrame的列的名字，而直接使用Axes绘图，参数更灵活
 5. 直接使用Axes绘图，可以非常方便的跟matplotlib结合使用
