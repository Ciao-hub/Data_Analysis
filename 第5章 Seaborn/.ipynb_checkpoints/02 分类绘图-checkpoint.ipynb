```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

**分类散点图**

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex')
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips)  # 点散开绘制
```

**箱线图**

```python
athletes = pd.read_csv('athlete_events.csv')
athletes.head()
```

```python
countries = {
    'CHN':'中国',
    'JPN':"日本",
    'KOR':'韩国',
    'USA':"美国",
    'CAN':"加拿大",
    'BRA':"巴西",
    'GBR':"英国",
    'FRA':"法国",
    'ITA':"意大利",
    'ETH':"埃塞俄比亚",
    'KEN':"肯尼亚",
    'NIG':"尼日利亚",
}
my_athletes = athletes[athletes['NOC'].isin(list(countries.keys()))]
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体
plt.figure(figsize = (10,5))
sns.boxplot(x = 'NOC', y = 'Height', data = my_athletes, hue = 'Sex')
plt.xticks(range(len(countries.values())),countries.values())  # 设置x轴坐标刻度和标签
plt.show()
```

**小提琴图**

```python
plt.figure(figsize = (10,5))
sns.violinplot(x = 'day', y = 'total_bill', data = tips,hue = 'sex', inner = 'box', split = True)
```

**条形图**

```python
# 统计平均数
plt.figure(figsize = (5,5))
sns.catplot(x = 'day', y = 'total_bill', kind = 'bar',data = tips, estimator = sum)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 统计比例
plt.figure(figsize = (5,5))
sns.catplot(x = 'sex', y = 'survived',data = titanic,kind = 'bar')
```

**柱状图**

```python
sns.catplot(x = 'sex',data = titanic, kind = 'count')  # x和y参数只能传入一个
```

```python
sns.catplot(y = 'sex',data = titanic, kind = 'count')
```

**点线图**

```python
sns.catplot(x = 'sex', y = 'survived', data = titanic, kind = 'point', hue = 'class')
```

## 笔记：
### 一、分类散点图
1. striplot和swarmplot
2. swarmplot采用一定算法，可以让点不重叠
3. 分类散点图，特别是swarmplot，不适合数据量特别多的。

### 二、分布图
1. 箱线图：boxplot或者catplot(kind = 'box')，然后使用hue参数可以进行分组
2. 小提琴图：violinplot或者catplot(kind = 'violin')，小提琴图是有KDE曲线(核密度估计曲线)估算而成，inner参数可以控制小提琴中的图，可以为box,quartile,stick,point，可以使用hue参数分组，split = True将小提琴图分割开来进行展示

### 三、统计图
1. 条形图：barplot或者catplot(kind = 'bar')，会自动统计（平均数、比例），也可以通过'estimator'来修改统计函数
2. 柱状图：countplot或者catplot(kind = 'count')，只能统计某个变量数据的个数，x，y只能传一个。
3. 点线图：pointplot或者catplot(kind = 'point")，可以看出某个变量变化关系
