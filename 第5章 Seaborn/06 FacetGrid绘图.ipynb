```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
g = sns.FacetGrid(tips)
g.map(plt.scatter,'total_bill','tip')
```

### 绘制多个图形

```python
g2 = sns.FacetGrid(tips,col = 'day',hue = 'time')  # 列 = 'day'，行 = 'time'
g2.map(plt.scatter,'total_bill','tip')
g2.add_legend()  # 添加图例
```

### 设置每个图形的尺寸

```python
g3 = sns.FacetGrid(tips,col = 'smoker',
                   height = 4,aspect = 1.2)   # height 表示每个图形尺寸，aspect 表示 '宽度/高度' 比例
g3.map(sns.barplot,'day','total_bill',order = ['Thur','Fri','Sat','Sun'])
```

### 设置图例

```python
sns.set(rc = {'font.sans-serif':'SimHei'})
g4 = sns.FacetGrid(tips,col = 'day',hue = 'time')
g4.map(plt.scatter,'total_bill','tip')
g4.add_legend(title = "标题",label_order = ['Lunch','Dinner'])
new_labels = ['午餐','晚餐']
for text,label in zip(g4._legend.texts,new_labels):
    text.set_text(label)
```

### 设置标题

```python
g5 = sns.FacetGrid(tips,col = 'day',row = 'time')
g5.map(sns.regplot,'total_bill','tip')
g5.set_titles(template = '时间{row_name}/星期{col_name}')
```

### 设置坐标轴

```python
g6 = sns.FacetGrid(tips,col = 'day',row = 'time')
g6.map(sns.scatterplot,'total_bill','tip')
g6.set_axis_labels('消费金额','小费')
g6.set(facecolor = 'skyblue')
```

### fig对象

```python
g7 = sns.FacetGrid(tips,col = 'day',row = 'time')
g7.map(sns.scatterplot,'total_bill','tip')
g7.fig.set_dpi(100)
```
