```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

```python
# 创建DataFrame最常见的方法是直接传入一个由等长列表或Numpy数组组成的字典
data = {
    'state':['Ohio','Ohio','Ohio','Nevade','Nevade','Nevade'],
    'year':[2000,2001,2002,2001,2002,2003],
    'pop':[1.5,1.7,3.6,2.4,2.9,3.2]
}
frame = pd.DataFrame(data)
frame
```

```python
# head选取前五行
frame.head()
```

```python
# 指定顺序排列
pd.DataFrame(data,columns = ['year','state','pop'])
```

```python
frame2 = pd.DataFrame(data,columns = ['year','state','pop','debt'],
                     index = ['one','two','three','four','five','six'])
frame2
```

```python
# 索引列获取为一个Series
print(frame2['state'])
print('\n')
print(frame2.year)
```

```python
# 获取行
frame2.loc['three']
```

```python
# 赋值
frame2['debt'] = 16.5
frame2
```

```python
frame2['debt'] = np.arange(6.)
frame2
```

```python
val = pd.Series([-1.2,-1.5,-1.7],index = ['two','four','five'])
frame2['debt'] = val
frame2
```

```python
frame2['eastern'] = frame2.state == 'Ohio'  # 添加一个布尔值的列，判断state是否为Ohio
frame2
```

```python
# del 删除
del frame2['eastern']
frame2.columns
```

```python
# 嵌套字典构建DataFrame
pop = {
    'Nevade':{2001:2.4,2002:2.9},
    'Ohio':{2000:1.5,2001:1.7,2002:3.6}
      }
frame3 = pd.DataFrame(pop)
frame3
```

```python
# 转置
frame3.T
```

```python
pd.DataFrame(pop,index = [2000,2001,2002])
```

```python
frame3
```

```python
pdata = {
    'Ohio':frame3['Ohio'][:-1],
    'Nevade':frame3['Nevade'][:2]
}
pd.DataFrame(pdata)
```

```python
# 设置index和columns的值
frame3.index.name = 'year';frame3.columns.name = 'state'
frame3
```

```python
# values属性
print(frame3.values)
print('\n')
print(frame2.values)
```
