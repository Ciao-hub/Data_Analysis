**读取数据**

```python
nations = pd.read_excel('nations.xlsx')
golds = pd.read_excel('golds.xlsx')
silvers = pd.read_excel('silvers.xlsx')
bronzes = pd.read_excel('bronzes .xlsx')

print("数据集nations：")
# print(type(notions))
nations.head()
```

concat官方文档：https://pandas.pydata.org/docs/reference/api/pandas.concat.html?highlight=concat#

```python
detailed = pd.concat([golds,silvers,bronzes])

print("数据集detailed：")
detailed.sample(5)
```

**数据预处理**

```python
type(detailed['名次'])
```

```python
detailed['名次'] = detailed['名次'].astype('str')
```

**缺失值**

```python
nations.isna()
```

```python
detailed.isna().sum().sum()   # 计算个数，第一个sum()显示每列的缺失值，第二个sum()显示总数
```

```python
def na_check(df):
    if df.isna().sum().sum() == 0:
        return "无缺失值"
    else:
        na_num = df.isna().sum().sum()
        df.dropna(axis = 0 , inplace = True)   # axis=0：处理行；inplace：用处理后的代替原DataFrame
        return "含有缺失值，已处理" + str(na_num) + "个缺失值"
```

```python
na_check(nations)
```

```python
na_check(detailed)
```

**重复值**

```python
def dup_check(df):
    if df.duplicated().sum() == 0:
        return '无重复值'
    else:
        dup_num = df.duplicated().sum()
        df.drop_duplicated(axis = 0, inplace = True)
        return '含有重复值，已处理' + str(dup_num) + '个重复值'
```

```python
dup_check(nations)
```

```python
dup_check(detailed)
```

**可视化**

```python
# 国家名称中英文映射
with open('D:\\Workspace\\Data_Analysis\\项目实战\\01 奥运冠军可视化\\country_ch_to_en.json') as f:
    nameMap = json.load(f)
    
nations = nations.replace(nameMap)
nations.replace('俄罗斯奥运队','Russia',inplace = True)
```

```python
attr = nations['国家'].tolist()  # 把Series返回一个列表
values = nations['总数'].tolist()

map_ = (
    Map()
    .add('总数',[list(z) for z in zip(attr,values)], 'world')
    .set_series_opts(label_opts = opts.LabelOpts(is_show = False))
    .set_global_opts(
        title_opts = opts.TitleOpts(title = '奥运会奖牌总数世界国家分布图', pos_left = 'center'),
        visualmap_opts = opts.VisualMapOpts(max_ = 120,is_piecewise = True,split_number = 8),
        legend_opts = opts.LegendOpts(is_show = False),
)
)

map_.load_javascript()
```

```python
map_.render_notebook()
```

**前三国家奖牌对比图**

```python
data = [
             {"name": "美国",
              "children": [
                  {"name": "金牌", "value": 39},
                  {"name": "银牌", "value": 41},
                  {"name": "铜牌", "value": 33},
              ]},
             {"name": "中国",
              "children": [
                      {"name": "金牌", "value": 38},
                      {"name": "银牌", "value": 32},
                      {"name": "铜牌", "value": 18},
              ]},
             {"name": "俄罗斯",
              "children": [
                      {"name": "金牌", "value": 20},
                      {"name": "银牌", "value": 28},
                      {"name": "铜牌", "value": 23},
              ]},
     ]
data
```

```python
sunburst = (
    Sunburst().add('',data_pair = data, radius=[0,'80%'])
    .set_global_opts(
        title_opts = opts.TitleOpts(title = "奥运会奖牌总数前三奖牌组成旭日图",pos_left = 'center'),
        legend_opts = opts.LegendOpts(is_show = True)
    )
    .set_series_opts(label_opts = opts.LabelOpts(formatter="{b}: {c}"))
)
sunburst.render_notebook()
```
