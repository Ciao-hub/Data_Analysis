### 统计各省大学数量

```python
# groupby详解：https://zhuanlan.zhihu.com/p/101284491
g = df.groupby('省市')  # groupby函数：将原有的DataFrame按照字段拆分成若干个分组DataFrame
```

```python
type(g)
```

```python
df_counts = g.count()['排名']
df0 = df_counts.copy()
df0.sort_values(ascending = False, inplace = True)
df0
```

### 各省大学平均分排序

```python
df_means0 = g.mean()['总分']
df_means = df_means0.round(2)  # round(2) 保留2位小数
df1 = pd.concat([df_counts,df_means], join = 'outer',  axis = 1)   # concat
df1.columns = ['数量','平均分']
df1.sort_values(by = ['平均分'], ascending = False, inplace = True)
df1
```

### 各省市大学数量和平均分柱状图

```python
df1.sort_values(by = ['平均分'], ascending = False, inplace = True)
d1 = df1.index.tolist()
d2 = df1['数量'].values.tolist()
d3 = df1['数量'].values.tolist()


bar0 = (
    Bar()
    .add_xaxis(d1)
    .add_yaxis('数量',d2)
    .add_yaxis('平均分数',d3)
    .set_global_opts(
        title_opts = opts.TitleOpts(title = '中国大学排名'),
        yaxis_opts = opts.AxisOpts(name = '量'),
        xaxis_opts = opts.AxisOpts(name = '省份')
    )
)
bar0.load_javascript()
```

```python
bar0.render_notebook()
```

### 各省市大学数量地图分布

```python
name = df0.index.tolist()  # 省份
count = df0.values.tolist()  # 各省的数量
m = (
    Map()
    .add('', [list(z) for z in zip(name,count)],'china')
    .set_global_opts(
        title_opts = opts.TitleOpts(title = '中国大学排名'),
        visualmap_opts = opts.VisualMapOpts(max_ = 40, split_number=8, is_piecewise=True)
    )
)

m.load_javascript()
```

```python
m.render_notebook()
```
